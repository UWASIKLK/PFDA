## Using matplotlib

Below are examples of how to use matpolotlib

### Import the relevant packages

`matplotlib.pyplot`

and

`numpy`

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Data input / creation

Creation of some data

`np.arange()` - this will create a series of data

```python
x = 0.3 + np.arange(10)
y = [1.5, 8.9, 4.3, 2.5, 5.9, 7.8, 6.1, 3.2, 1.1, 9.5]
y2 = [5.5, 7.3, 4.3, 3.5, 1.9, 6.2, 4.8, 7.2, 1.9, 9.5]
```

### Marking a plot

we have below functions for creating a plot

`fig, ax = plt.subplots()`- this will create a plot

and

`plt.show()`- this will display the plot

```python
fig, ax = plt.subplots()
plt.show()

#this will create basic plot as per below
```

### Setting paraments

`set()`

in this function we can specify different set ups together
the basic set up will be for label (`xlabel, ylabel`), limitation (`xlim`) and title (`title`)

```python
fig, ax = plt.subplots()

ax.bar(x,y)
#ax.plot(x,y)
ax.set (xlabel= "amount", ylabel="period", title="amount vs period")

plt.show()
```

or we can use individual functions for each parameter separately

`set_title, set_xlabel, set_ylabel`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
#ax.plot(x,y)
ax.set_title("amount vs period")
ax.set_xlabel("amount")
ax.set_ylabel("period")
plt.show()
```

### Formating the labels

We will use formatter function `ax.xaxis.set_major_formatter(formatter_fun)` to format the labels

But we will need to create the funcition that will do the formatting first
This function will take two parameters, value on the axis and position of the tick
It will then return the text 

```python
def currency_formatter(x,pos):
    s = f"€{x}"
    return s
```

Once the function is created we need to pass it to the `ax.xaxis.set_major_formatter()`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
#ax.plot(x,y2)

ax.set_title("amount vs period")
ax.set_xlabel("amount")
ax.set_ylabel("period")

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')


plt.show()
```

### Adding annotation with text

To add text with the annotation we need to use function `ax.annotate('text', (x_pointer, y_pointer), xytext = (x,y))`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
#ax.plot(x,y2)

ax.set_title("amount vs period")
ax.set_xlabel("amount")
ax.set_ylabel("period")

ax.annotate('maximum', (9, 9.5), xytext=(6, 9.5
                                         ),arrowprops=dict(facecolor='green', shrink=0.01))

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')


plt.show()
```
