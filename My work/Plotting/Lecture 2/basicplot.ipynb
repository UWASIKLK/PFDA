# Basic Plot

Imports of the appropriate packages

```python
import matplotlib.pyplot as plt
import numpy as np
```

Bar (x, height)
Add a data

```python


x = 0.5 + np.arange (10)
y = [1.7, 5.6, 2.3, 4.2, 0.8, 9.4, 7.7, 8.1, 3.5, 5.9]

#plot
fig, ax = plt.subplots()

ax.bar (x,y, width=1, edgecolor="white", linewidth = 0.7)

ax.set (xlim=(0,10), xticks = np.arange (1,10),
        ylim=(0,10), yticks = np.arange (1,10))

plt.style.use('classic')
plt.show()

```

Plot(x,y)

```python
plt.style.use('fivethirtyeight')

x = np.linspace(0, 20, 90)
y = 4.5 + 1 * np.sin(1 * x)
x2 = np.linspace(0, 20, 45)
y2 = 4.5 + 1 * np.sin(1 * x2)

# plot
fig, ax = plt.subplots()

ax.plot(x2, y2 + 2.5, 'x', markeredgewidth=2)
ax.plot(x, y, linewidth=2.0)
ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)

ax.set(xlim=(0, 9), xticks=np.arange(1, 9),
       ylim=(0, 9), yticks=np.arange(1, 9))

plt.show()
```

scatter(x,y)

```python
plt.style.use('seaborn-bright')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

Fill in between example


```python
N = 21
x = np.linspace(0, 10, 11)
y = [3.9, 4.4, 10.8, 10.3, 11.2, 13.1, 14.1,  9.9, 13.9, 15.1, 12.5]

# fit a linear curve and estimate its y-values and their error.
a, b = np.polyfit(x, y, deg=1)
y_est = a * x + b
y_err = x.std() * np.sqrt(1/len(x) +
                          (x - x.mean())**2 / np.sum((x - x.mean())**2))

fig, ax = plt.subplots()
ax.plot(x, y_est, '-')
ax.fill_between(x, y_est - y_err, y_est + y_err, alpha=0.2)
ax.plot(x, y, 'o', color='tab:green')

plt.show()
```
