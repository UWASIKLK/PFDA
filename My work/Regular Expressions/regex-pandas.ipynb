## Using regex in Pandas

Import libraries and data from csv file

```python
import pandas as pd
```

```python
datadir = "./data/"
filename ="people-100.csv"

people = pd.read_csv(datadir + filename) 
people.head(5)                           
```

### Split the email address

I want to split the email address to get the domains

```python
people ['Domain'] = people['Email'].str.split('@').str[1]
people.head(5)
```

This is different way how to do it using the regular expressions.

```python
pattern = r".*@"  #letter 'r' means that this is regular rexpression
people ['Domain'] = people['Email'].str.replace(pattern, '', regex=True)
people.head(3)
```

This is more preceise pattern to use.

```python
pattern = r".*@([\w\.]+\.\w{2,3})"
people['Domain'] = people['Email'].str.replace(pattern, '\\1', regex=True)
people.head(3)
```

## Phone numbers

I will assume that all numbers are phone numbers.

I will split the number - anything prior extention will go to one column and extention numbers will go to another column.

```python
extention = r"x\d*" #r is for regular expression, 'x' is the separator for number and extention, \d* means any number after the x - this is what I want to replace (x including any number after it)
people ['clean_phone'] = people['Phone'].str.replace(extention, '', regex=True) #this create new column 'clean_phone' where any numbers after 'x' are replaced by blank '' 
people.head(5)
```

```python
number = r"[\.\- \(\)]*" #I want to replace anything which contains .,-, ,() with blank ''. All special characters as separated by backsplash '\'
people['clean_phone'] = people['clean_phone'].str.replace(number,'', regex=True) #in same column 'clean_phone' new phone number withouht any .,-, ,() will be placed
people.head(10)
```

Some of the numbers has + or 00 in front of the actual phone number.

I will replace all numbers having 00 in front with +1

```python
prefix = r"^([^\+0].*)" #this will find numbers which don't start with either + or 0, then take all there rest characters after this (which is as (.*))
people['clean_phone'] = people['clean_phone'].str.replace(prefix, '+1\\1', regex=True) #this will replace nubmer which don't have + or 0 with the +1 character follow by rest of the original numbers (this is as \\1)
people.head(10)
```

There are still numbers which starts with '00'.

I will find them and replace them with +.

```python
prefix = r"^00(.*)" #this will find numbers which starts with '00'
people ['clean_phone'] = people ['clean_phone'].str.replace(prefix, '+1\\1', regex=True) #this will replace'00' with '+1' following by the rest of the numbers
people.head(10)
```

I need to create new column which will have extention numbers only.

```python
extension=r"^.*x" #this will find all numbers which has extention number after 'x'
people['extension'] = people['Phone'].str.replace(extension, '', regex=True) #this will take the extention and create separate column
people.head(10)
```

This will extract extention number.

```python
extension=r"x(\d+)" #this will find numbers which has x and any numbers after
people['extension'] = people['Phone'].str.extract(extension) #this will extract extension from the 'Phone' numbers and put the result to 'extension' column
people.head(10)                                              #if the phone number has no extenntion (x following by nunbers) it will put NaN in extension column
```
