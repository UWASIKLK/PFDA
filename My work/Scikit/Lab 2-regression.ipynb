### Lab 2 - regression

1. We will write a program which will load the seaborn 'tips' data and prints out the top 5 rows

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

#load the dataset
dataset =  sns.load_dataset('tips')
print(dataset.head(5))
```

```python
#creating the regression plot for 'tips' against the 'total_bill' 
sns.set_theme(color_codes=True)
sns.lmplot(x='total_bill', y='tip', order= 1, data=dataset)
plt.show()
```

```python
#plotting the size and tips with the third variable = smoker
# map 'hue' levels to colours with a dictionary
sns.lmplot(x='size', y='tip', hue='smoker', palette=dict(Yes="g", No="m"),data=dataset)
plt.show()
```

```python
sns.lmplot(x='size', y='tip',data=dataset, hue = 'sex', palette= 'cubehelix', x_jitter=.05) #we put a jitter to make it easier to see and with third variable gender
plt.show()
```

```python
#finding the mean tip for each size
sns.lmplot(x='size', y='tip', data=dataset, x_estimator= np.mean)
plt.show()
```
