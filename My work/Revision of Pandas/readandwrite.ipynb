## Reading and Writing in Pandas

In this notebook I will have a look how to write and read from documents using pandas

### Import 

Import necessary libraries and load the data from the directory and filename

```python
import pandas as pd

#checking if the file and directory are there

!dir "./data"
```

```python
datadir = "./data/"
filename ="people-100.csv"
```

### Read in the csv file

```python
people = pd.read_csv(datadir + filename)  # reading in the file
people.head(5)                            # checking first 5 lines
```

```python
people.info()
```

```python
people.sort_values(by ="Sex") #sorting by gender
```

```python
people.columns # list of all columns
```

```python
name = people.loc[:, ["First Name", "Last Name"]] #from whole data it will select only First & Last name column
name.head(10)                                     #this will print out first 10 rows
```

```python
name.iloc[5:10] #this will print out only selected rows 5-10
```

If I want to display only certain colums, I need to use `header = [list of cols]`

```python
columns = ['First Name', 'Last Name', 'Email', 'Sex', 'Date of birth']
people = pd.read_csv (datadir + filename, usecols =columns)
people.head(5)
```

The **'Date of birth'** is an object and I want to change it to the datetime object.

```python
people.info(['Date of birth'])
```

To do so I need to use `pd.read_csv(file, parse_date = [column])`

```python
people = pd.read_csv(datadir + filename, parse_dates=['Date of birth'])

print (people ['Date of birth']) 
```

I want to change the format of the "**Date of birth**" column to `date-month-year`

```python

people = people['Date of birth'].dt.strftime("%d-%m-%Y")
people.head(10)
```
